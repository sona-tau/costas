```julia
function fromtxt(s::String)::Vector{Vector{Int}}
    map.(Base.Fix1(parse, Int), split.(eachline(s)))
end
```

```julia
function totxt(c::Vector{Vector{Integer}})::String
    join(join.(c, " "), "\n")
end
```

```julia
fromtxt("../../data/costas_10x10.txt")
```

```julia
function convert(fpath::String)
    c = loadvar(fpath)
    println("$fpath: $(typeof(c))")
    text = totxt(c)
    open(fpath[1:end - 3] * "txt", "w") do file
        write(file, text)
    end
end
```
