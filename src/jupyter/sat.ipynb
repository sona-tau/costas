```python
f = Function("f", IntSort(), IntSort())
```

```python
n = 5
```

```python
s = Solver()
n = Int("n")
i = Int("i")
j = Int("j")
k = Int("k")
s.add(n == 5)
s.add(1 <= k)
s.add(k <= n - 1)
s.add(0 <= i)
s.add(i <= n - k)
s.add(0 <= j)
s.add(j <= n - k)
s.add(Implies(f(i + k) - f(i) == f(j + k) - f(j), i == j))
```

```python
s.check()
```

```python
s.model()
```
