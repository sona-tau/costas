```julia
function checkwelch(costas::Vector{Vector{Integer}}, prime::Integer)
    all(all(w == lst || isrotated(w, lst) || isrotated(w, reverse(lst)) for w in welch(prime)) for lst in costas)
end
```

```julia
println("From costas_2x2:", checkwelch(costas_2x2,3))
println("From costas_4x4:", checkwelch(costas_4x4,5))
println("From costas_6x6:", checkwelch(costas_6x6,7))
println("From costas_10x10:", checkwelch(costas_10x10,11))
```

```julia
function filterwelchs(costas, prime)
    filter(lst -> all(!isrotated(lst, w) && !isrotated(lst, reverse(w)) for w in welch(prime)), costas)
end
filterwelchs(costas_4x4, 5)
```

```julia
# /*
#  * Complexity: O(n^3)
#  */
# template <class T> fn build_all_naive(Vec<T> const& vec, T x) -> Vec<Vec<T>> {
#     var out = Vec<Vec<T>>();
#     for (Size i = 0; i < vec.size(); ++i) {
#         var lst2 = vec;
#         lst2.insert(lst2.begin() + i, x);
#         if (is_costas(lst2))
#             out.push_back(lst2);
#     }
#     return out;
# }

function buildallnaive(vec::Vector{T}, x::T)::Vector{Vector{T}} where T
    out = []
    for i in 1:length(vec)
        lst2 = copy(vec)
        insert!(lst2, i, x)
        if iscostas(lst2)
            push!(out,lst2)
        end
    end
    out
end
```

```julia
function iscostashelper2(lst::Vector{T}, f::Function, n::Integer)::Bool where T
    all(begin
        seen::Vector{Bool} = zeros(Bool, 2 * n)
        all(begin
                result = f(lst[i+k],lst[i],n)
                prev = seen[result]
                seen[result] = true
                !prev
            end for i ∈ 1:(n - k))
        end for k ∈ 1:n)
end
```

```julia
function ts(n::Integer)::Vector{Vector{Integer}}
    out = [[1]]
    for i ∈ 1:n
        new_out = []
        for vec ∈ out
            println(vec)
            verts = findverts(vec)
            println(verts)
            for vert ∈ verts
                c = swapidx([copy(vec);[i+1]], vert, i)
                push!(new_out, c)
                println("$new_out, $i")
            end
            push!(new_out, [copy(vec);[i+1]])
        end
        new_out
        out = new_out
    end
    out
end
```

```julia
function findverts(vec::Vector)::Vector
    out = []
    for i in 1:length(vec)
        if vec[i] == i
            push!(out, i)
        end
    end
    out
end
```

```julia
function swapidx(vec::Vector{T}, i::Integer, j::Integer)::Vector{T} where T
    tmp = vec[i]
    vec[i] = vec[j]
    vec[j] = tmp
    vec
end
```

```julia
function isprimitiveroot(prime::Integer, alpha::Integer)::Bool 
    not_one::Bool = all(powermod(alpha, i, prime) != prime - 1 for i in 1:((prime - 1) ÷ 2 - 1))
    last_one::Bool = powermod(alpha, (prime - 1) ÷ 2, prime) == prime - 1
    not_one && last_one
end
```

```julia
struct Unit
end

struct Pair{T, S}
    l::T
    r::S
    function fst()::T
        l
    end
    function snd()::S
        r
    end
end

# Check if a sequence is a Costas sequence with debug information
function iscostasdbg(lst::Vector, f::Function = normal_predicate) #::Union{Unit, Tuple{Int64, Int64}}
    n::Integer = length(lst)
    for k in 1:n
        diffs = [(false, (-1,-1)) for _ in 1:2 * n]
        for i in 1:(n - k)
            a = lst[i + k]
            b = lst[i]
            result = f(a, b, n)
            if diffs[result][1]
                return ((i, i + k), diffs[result][2])
            elseif result != n
                diffs[result] = (true, (i, i + k))
            end
        end
    end
    return Unit()
end
```

```julia
using Base.Threads
function searchdbg(n::Integer)::Vector{Vector{Integer}}
    p1 = 0
    p2 = 0
    p3 = 0
    p4 = 0
    i1 = 0
    i2 = 0
    i3 = 0
    i4 = 0
    acc::Vector{Vector{Vector{Integer}}} = [[] for _ in 1:nthreads()]
    
    @threads :static for i in 1:factorial(n)
        tid = threadid()
        lst = nthperm(1:n, i)
        if (i1 != 0 && p1 == lst[i1]) && (i2 != 0 && p2 == lst[i2]) && (i3 != 0 && p3 == lst[i3]) && (i4 != 0 && p4 == lst[i4])
            #println("miss: ", lst)
            continue
        end
        result = iscostasdbg(lst)
        if result isa Unit
            push!(acc[tid], lst)
        else
            #println("fail: ", lst, "; ", result)
            (i1, i2), (i3, i4) = result
            p1, p2, p3, p4 = lst[i1], lst[i2], lst[i3], lst[i4]
        end
    end
    reduce(vcat, acc)
end
```

```julia
using Base.Threads
function searchone(n::Integer)::Union{Nothing, Vector{Integer}}
    p1 = 0
    p2 = 0
    p3 = 0
    p4 = 0
    i1 = 0
    i2 = 0
    i3 = 0
    i4 = 0
    
    for i in 1:factorial(n)
        lst = nthperm(1:n, i)
        if (i1 != 0 && p1 == lst[i1]) && (i2 != 0 && p2 == lst[i2]) && (i3 != 0 && p3 == lst[i3]) && (i4 != 0 && p4 == lst[i4])
            continue
        end
        result = iscostasdbg(lst)
        if result isa Unit
            return lst
        else
            (i1, i2), (i3, i4) = result
            p1, p2, p3, p4 = lst[i1], lst[i2], lst[i3], lst[i4]
        end
    end
    return nothing
end
```

```julia
function lempelc(prime::Integer, c::Integer = 1)::Vector{Vector{Integer}}
    welch(prime) .|> lst -> init(lst) .|> n -> findindex(lst, prime + c - n)
end
```
