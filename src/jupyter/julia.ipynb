```julia
IntOrChar = Union{Int64, AbstractChar}

function default(x, y, n)
    x - y + n
end

function iscostas(lst::Vector, f::Function = default)::Bool
    if size(unique(lst), 1) != size(lst, 1)
        return false
    end
    n = size(lst, 1)
    flag = false
    for k = 2:n
        diffs = [false for _ = 1:(2 * n - 1)]
        if flag
            continue
        end
        for i = 1:(n - k + 1)
            a = lst[(i + k) % n + 1]
            b = lst[i]
            result = f(a, b, n)
            if diffs[result]
                return false
            elseif result != n
                diffs[result] = true
            end
            flag = diffs[result]
        end
    end
    return flag
end
```

```julia
iscostas([1,2,4,3])
```

```julia
predicate = (x, y, n) -> x == '*' || y == '*' ? n : x - y + n
```

```julia
iscostas([1,2,'*',3], predicate)
```
