```julia
function checkwelch(costas::Vector{Vector{Integer}}, prime::Integer)
    all(all(w == lst || isrotated(w, lst) || isrotated(w, reverse(lst)) for w in welch(prime)) for lst in costas)
end
```

```julia
println("From costas_2x2:", checkwelch(costas_2x2,3))
println("From costas_4x4:", checkwelch(costas_4x4,5))
println("From costas_6x6:", checkwelch(costas_6x6,7))
println("From costas_10x10:", checkwelch(costas_10x10,11))
```

```julia
function filterwelchs(costas, prime)
    filter(lst -> all(!isrotated(lst, w) && !isrotated(lst, reverse(w)) for w in welch(prime)), costas)
end
filterwelchs(costas_4x4, 5)
```

```julia
# /*
#  * Complexity: O(n^3)
#  */
# template <class T> fn build_all_naive(Vec<T> const& vec, T x) -> Vec<Vec<T>> {
#     var out = Vec<Vec<T>>();
#     for (Size i = 0; i < vec.size(); ++i) {
#         var lst2 = vec;
#         lst2.insert(lst2.begin() + i, x);
#         if (is_costas(lst2))
#             out.push_back(lst2);
#     }
#     return out;
# }

function buildallnaive(vec::Vector{T}, x::T)::Vector{Vector{T}} where T
    out = []
    for i in 1:length(vec)
        lst2 = copy(vec)
        insert!(lst2, i, x)
        if iscostas(lst2)
            push!(out,lst2)
        end
    end
    out
end
```

```julia
function iscostashelper2(lst::Vector{T}, f::Function, n::Integer)::Bool where T
    all(begin
        seen::Vector{Bool} = zeros(Bool, 2 * n)
        all(begin
                result = f(lst[i+k],lst[i],n)
                prev = seen[result]
                seen[result] = true
                !prev
            end for i ∈ 1:(n - k))
        end for k ∈ 1:n)
end
```

```julia
function ts(n::Integer)::Vector{Vector{Integer}}
    out = [[1]]
    for i ∈ 1:n
        new_out = []
        for vec ∈ out
            println(vec)
            verts = findverts(vec)
            println(verts)
            for vert ∈ verts
                c = swapidx([copy(vec);[i+1]], vert, i)
                push!(new_out, c)
                println("$new_out, $i")
            end
            push!(new_out, [copy(vec);[i+1]])
        end
        new_out
        out = new_out
    end
    out
end
```

```julia
function findverts(vec::Vector)::Vector
    out = []
    for i in 1:length(vec)
        if vec[i] == i
            push!(out, i)
        end
    end
    out
end
```

```julia
function swapidx(vec::Vector{T}, i::Integer, j::Integer)::Vector{T} where T
    tmp = vec[i]
    vec[i] = vec[j]
    vec[j] = tmp
    vec
end
```

```julia
function isprimitiveroot(prime::Integer, alpha::Integer)::Bool 
    not_one::Bool = all(powermod(alpha, i, prime) != prime - 1 for i in 1:((prime - 1) ÷ 2 - 1))
    last_one::Bool = powermod(alpha, (prime - 1) ÷ 2, prime) == prime - 1
    not_one && last_one
end
```

```julia
struct Unit
end

struct Pair{T, S}
    l::T
    r::S
    function fst()::T
        l
    end
    function snd()::S
        r
    end
end

# Check if a sequence is a Costas sequence with debug information
function iscostasdbg(lst::Vector, f::Function = normal_predicate) #::Union{Unit, Tuple{Int64, Int64}}
    n::Integer = length(lst)
    for k in 1:n
        diffs = [(false, (-1,-1)) for _ in 1:2 * n]
        for i in 1:(n - k)
            a = lst[i + k]
            b = lst[i]
            result = f(a, b, n)
            if diffs[result][1]
                return ((i, i + k), diffs[result][2])
            elseif result != n
                diffs[result] = (true, (i, i + k))
            end
        end
    end
    return Unit()
end
```

```julia
using Base.Threads
function searchdbg(n::Integer)::Vector{Vector{Integer}}
    p1 = 0
    p2 = 0
    p3 = 0
    p4 = 0
    i1 = 0
    i2 = 0
    i3 = 0
    i4 = 0
    acc::Vector{Vector{Vector{Integer}}} = [[] for _ in 1:nthreads()]
    
    @threads :static for i in 1:factorial(n)
        tid = threadid()
        lst = nthperm(1:n, i)
        if (i1 != 0 && p1 == lst[i1]) && (i2 != 0 && p2 == lst[i2]) && (i3 != 0 && p3 == lst[i3]) && (i4 != 0 && p4 == lst[i4])
            #println("miss: ", lst)
            continue
        end
        result = iscostasdbg(lst)
        if result isa Unit
            push!(acc[tid], lst)
        else
            #println("fail: ", lst, "; ", result)
            (i1, i2), (i3, i4) = result
            p1, p2, p3, p4 = lst[i1], lst[i2], lst[i3], lst[i4]
        end
    end
    reduce(vcat, acc)
end
```

```julia
# Check if a sequence is a Costas sequence with debug information
function iscostas(lst::SVector, f::Function = normal_predicate)::Union{Unit, Tuple{Tuple{Int64, Int64}, Tuple{Int64, Int64}}}
    n::Int64 = length(lst)
    for k in 1:n
        diffs::Vector{Tuple{Bool, Tuple{Int64, Int64}}} = [(false, (-1,-1)) for _ in 1:2 * n]
        for i in 1:(n - k)
            a = lst[i + k]
            b = lst[i]
            result = f(a, b, n)
            if diffs[result][1]
                return ((i, i + k), diffs[result][2])
            elseif result != n
                diffs[result] = (true, (i, i + k))
            end
        end
    end
    return Unit()
end
```

```julia
using Base.Threads
function searchallt(n::Int64)::Vector{SVector{n, Int64}}
    p1::Int64 = 0
    p2::Int64 = 0
    p3::Int64 = 0
    p4::Int64 = 0
    i1::Int64 = 0
    i2::Int64 = 0
    i3::Int64 = 0
    i4::Int64 = 0
    acc::Vector{Vector{SVector{n, Integer}}} = [[] for _ in 1:nthreads()]

    @threads :static for i in 1:factorial(n)
        tid::Int64 = threadid()
        lst::SVector{n, Int64} = nthperm(1:n, i)
        if (i1 != 0 && p1 == lst[i1]) && (i2 != 0 && p2 == lst[i2]) && (i3 != 0 && p3 == lst[i3]) && (i4 != 0 && p4 == lst[i4])
            continue
        end
        result = iscostas(lst)
        if result isa Unit
            push!(acc[tid], lst)
        else
            (i1, i2), (i3, i4) = result
            p1, p2, p3, p4 = lst[i1], lst[i2], lst[i3], lst[i4]
        end
    end
    reduce(vcat, acc)
end
```

```julia
function searchall(n::Int64)::Set{Vector{Int64}}
    p1::Int64 = 0
    p2::Int64 = 0
    p3::Int64 = 0
    p4::Int64 = 0
    i1::Int64 = 0
    i2::Int64 = 0
    i3::Int64 = 0
    i4::Int64 = 0
    acc::Set{Vector{Int64}} = Set()

    for i in 1:(factorial(n) ÷ 2)
        lst::Vector{Int64} = nthperm(1:n, i)
        if (i1 != 0 && p1 == lst[i1]) && (i2 != 0 && p2 == lst[i2]) && (i3 != 0 && p3 == lst[i3]) && (i4 != 0 && p4 == lst[i4])
            continue
        end
        result = iscostas(lst)
        if result isa Unit
            push!(acc, lst)
            push!(acc, lst |> reverse)
            lst2 = lst .|> x -> (-x % n) + n
            #push!(acc, lst2)
            #push!(acc, lst2 |> reverse)
        else
            (i1, i2), (i3, i4) = result
            p1, p2, p3, p4 = lst[i1], lst[i2], lst[i3], lst[i4]
        end
    end
    acc
end

```

```julia
using Base.Threads
function searchone(n::Integer)::Union{Nothing, Vector{Integer}}
    p1 = 0
    p2 = 0
    p3 = 0
    p4 = 0
    i1 = 0
    i2 = 0
    i3 = 0
    i4 = 0
    
    for i in 1:factorial(n)
        lst = nthperm(1:n, i)
        if (i1 != 0 && p1 == lst[i1]) && (i2 != 0 && p2 == lst[i2]) && (i3 != 0 && p3 == lst[i3]) && (i4 != 0 && p4 == lst[i4])
            continue
        end
        result = iscostasdbg(lst)
        if result isa Unit
            return lst
        else
            (i1, i2), (i3, i4) = result
            p1, p2, p3, p4 = lst[i1], lst[i2], lst[i3], lst[i4]
        end
    end
    return nothing
end
```

```julia
function seq(lst::Vector{<:Integer})::Vector{Integer}
    acc = []
    for i ∈ 1:length(lst)
        counter = 0
        for el ∈ lst
            if el < i
                counter += 1
            elseif el == i
                break
            end
        end
        push!(acc, counter)
    end
    acc
end
```

```julia
function unseq(lst::Vector{<:Integer})::Vector{Integer}
    acc = []
    for (idx, elem) ∈ enumerate(lst)
        l, r = splitat(acc, elem)
        acc = [l;[idx];r]
    end
    acc
end
```

```julia
function splitat(v::Vector{T}, i::Integer)::Tuple{Vector{T}, Vector{T}} where T
    (v |> Take(i), v |> Drop(i)) .|> collect
end
```

```julia
function dfs(current::Vector{<:Integer})::Vector{Integer}
    longest = copy(current)
    while length(current) != 0
        if iscostas(unseq(current))
            if length(current) > length(longest)
                longest = copy(current)
            end
            push!(current, 0)
        else
            while length(current) > 0 && last(current) + 1 >= length(current)
                pop!(current)
            end
            if length(current) == 0
                break
            end
            current[length(current)] = last(current) + 1
        end
    end
    longest
end
```

```julia

size = 4
lists = costas_4x4
last = lists[1][1:size - 3]
inputs = []
arr = []
for l in lists
    if last != l[1:size - 3]
        push!(inputs, arr)
        global last = l[1:size - 3]
        global arr = []
    end
    push!(arr, (l[size - 2], l[size - 1], l[size]))
end

println("inputs has $(length(inputs)) many elements")
println(inputs)

limit = length(inputs)
colors = palette([:red, :orange, :yellow, :green, :blue, :indigo, :violet], limit)
```

```julia
idx = 0
inputs = inputs .|> l -> begin
    global idx += 1
    (l, idx)
end
```

```julia
@gif for i in 0:360
    for pt in costas_3x3
        x = pt[1]
        y = pt[2]
        z = pt[3]
        scatter!(x, y, z; markercolor = colors[1], camera = (i, round(atand(1 / √2); digits = 3)), kw...)
    end
end
```
